```python
import pandas as pd
from datetime import date, timedelta
import yahooquery as yq
from tickers import tickers
```

```python
def allsundays(year):
   d = date(year, 1, 1)                    # January 1st
   d += timedelta(days = 11 - d.weekday())  # First Sunday
   while d.year == year:
      yield d
      d += timedelta(days = 7)
    
fridays = []
for d in allsundays(2022):
   fridays.append(d.isoformat())
```
